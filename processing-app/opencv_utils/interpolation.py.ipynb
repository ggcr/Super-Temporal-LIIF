```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

def disp(cv2_im):
    color = cv2.cvtColor(cv2_im, cv2.COLOR_BGR2RGB)
    plt.axis('off')
    plt.imshow(color)

# reads image 'opencv-logo.png' as grayscale
img = cv2.imread('./data/uab.jp2')

disp(img)
```

```python
b = cv2.resize(img, (img.shape[1] * 2,img.shape[0] * 2), cv2.INTER_LINEAR)
disp(b)
```

```python
c = cv2.resize(img, (img.shape[1] * 2,img.shape[0] * 2), cv2.INTER_NEAREST)
disp(c)
```

```python
d = cv2.resize(img, (img.shape[1] * 2,img.shape[0] * 2), cv2.INTER_CUBIC)
disp(d)
```
